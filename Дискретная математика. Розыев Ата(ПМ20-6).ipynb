# 1) Неориентированный граф с 100+ вершинами.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import networkx as nx
from itertools import combinations
from random import random


def ER(n, p):
    V = set([v for v in range(n)])
    E = set()
    for combination in combinations(V, 2):
        a = random()
        if a < p:
            E.add(combination)

    g = nx.Graph()
    g.add_nodes_from(V)
    g.add_edges_from(E)

    return g


n = 111 # Количество вершин
p = 0.00232
G = ER(n, p)
pos = nx.spring_layout(G)
nx.draw_networkx(G, pos)
plt.title("Граф")
plt.show()
```

# 2) Порядок обхода в ширину

```python
def dfs(graph, node, visited1):
    if node not in visited1:
        visited1.append(node)
        for n in graph[node]:
            dfs(graph,n, visited1)
    return visited1

visited1 = dfs(G,0, []) # В скобках после G указываете начальную точку([] не трогать!!!)
print(visited1)
```

Существует ли путь к вершине

```python
def retunShortestPath(G, start, end):

    queue = [(start,[start])]
    visited = set()

    while queue:
        n, p = queue.pop(0)
        visited.add(n)
        for node in G[n]:
            if node == end:
                return p + [end]
            else:
                if node not in visited:
                    visited.add(node)
                    queue.append((node, p + [node]))
print(retunShortestPath(G, 0,110)) # В скобках после G указываете начальную точку и конечную, если связи нет, то выдаст None
```
